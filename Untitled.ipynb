```python
%pylab inline
import networkx as nx
```

```python
undersea = nx.read_gpickle('data/undersea_today.networkx')
terrestrial = nx.read_gpickle('data/terrestrial_today.networkx')
```

```python
worldnet = nx.compose(undersea,terrestrial)
```

```python
nx.draw(worldnet, with_labels=True)
```

```python
# Next, use nx.connected_components to get the list of components,
# then use the max() command to find the largest one:
components = nx.connected_components(worldnet)
largest_component = max(components, key=len)

# Create a "subgraph" of just the largest component
# Then calculate the diameter of the subgraph, just like you did with density.
#

subgraph = worldnet.subgraph(largest_component)
diameter = nx.diameter(subgraph)
diameter
```

```python
from pycountry_convert import country_alpha2_to_country_name
from operator import itemgetter
import pandas as pd

def top_n (my_dict, n, reverse=True):
    sorted_dict = sorted(my_dict.items(), key=itemgetter(1), reverse=reverse)
    return sorted_dict[:n]

def plot_top (top_lst, title='Centrality', **kwargs):
    degree = pd.DataFrame(top_lst, columns=['Country', title])
    degree.index = degree['Country']
    degree['Country'] = degree['Country'].apply(country_alpha2_to_country_name)
    ax = degree.plot.bar(**kwargs)
    ax.set_xlabel('')
    return ax

degree_dict = dict(worldnet.degree(worldnet.nodes()))
nx.set_node_attributes(worldnet, degree_dict, 'degree')
top_degree = top_n(degree_dict, 20)
plot_top(top_degree)
```

```python
betweenness_dict = nx.betweenness_centrality(worldnet) # Run betweenness centrality
nx.set_node_attributes(worldnet, betweenness_dict, 'betweenness')
ax = plot_top(top_n(betweenness_dict, 14), title='Betweenness centrality', figsize=(12, 5),
             rot=45 
             )
```

```python
betweenness_dist = list(betweenness_dict.values())
betweenness_dist = [b for b in betweenness_dist if b >0]
```

```python
import powerlaw
results = powerlaw.Fit(betweenness_dist)
results.alpha, results.xmin
```

```python
from scipy.stats import kstest
kstest(betweenness_dist, 'powerlaw', args=(results.alpha, results.xmin), N=len(betweenness_dist))
```

```python
kstest(betweenness_dist, 'expon')
```

```python
from scipy.stats import anderson
anderson(betweenness_dist, dist='logistic')
```

```python
import geopandas
worldmap = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))

def plot (my_map, col, **kwargs):
    fig, ax = plt.subplots(1, 1)
    my_map.plot(column=col, ax=ax, legend=True, figsize=(20, 10), **kwargs)
    
plot(worldmap, 'pop_est')
```

```python
from pycountry_convert import country_name_to_country_alpha3

def marshall_graph_measure (graph_measure) -> pd.DataFrame:
    '''
    Turn our alpha2 graph measures into alpha3 series for geopandas.
    '''
    to_alpha3 = lambda alpha2:  country_name_to_country_alpha3(country_alpha2_to_country_name(alpha2))
    
    alpha3_dict = {}
    for country, val in betweenness_dict.items():
        try:
            c = to_alpha3(country)
            alpha3_dict[c] = val
        except:
            print('WARN: error parsing country!', country, 'value', val)
    return pd.DataFrame({'iso_a3': alpha3_dict.keys(), 'graph_measure': alpha3_dict.values()})

metric_df= marshall_graph_measure(betweenness_dict)
```

```python
worldmap_graphmeasure = worldmap.merge(metric_df, on='iso_a3')
```

```python
plot(worldmap_graphmeasure, 'graph_measure', cmap='PiYG');
```
