# Merge undersesa and terrestrial networks

```python
import pickle

def read_pickle (fn):
    with open(fn, 'rb') as handle:
        return pickle.load(handle)
```

```python
undersea = read_pickle('data/undersea_edges_today.pickle')
terrestrial = nx.read_gpickle('data/terrestrial_edges_today.pickle')
```

```python
edges = undersea.copy()

for key in terrestrial:
    if key in edges:
        edges[key] = terrestrial[key] + edges[key]
    else:
        pass
          
edges[('CA', 'US')]
```

```python
worldnet = nx.Graph()
for edge, weight in edges.items():
    a, b = edge
    worldnet.add_edge(a, b, weight=weight)
```

```python
worldnet.edges[('CN', 'JP')]
```

```python
# Next, use nx.connected_components to get the list of components,
# then use the max() command to find the largest one:
components = nx.connected_components(worldnet)
largest_component = max(components, key=len)

# Create a "subgraph" of just the largest component
# Then calculate the diameter of the subgraph, just like you did with density.
#

subgraph = worldnet.subgraph(largest_component)
diameter = nx.diameter(subgraph)
diameter
```

```python
from pycountry_convert import country_alpha2_to_country_name
from operator import itemgetter
import pandas as pd

def top_n (my_dict, n, reverse=True):
    sorted_dict = sorted(my_dict.items(), key=itemgetter(1), reverse=reverse)
    return sorted_dict[:n]

def plot_top (top_lst, title='Centrality', **kwargs):
    degree = pd.DataFrame(top_lst, columns=['Country', title])
    degree.index = degree['Country']
    degree['Country'] = degree['Country'].apply(country_alpha2_to_country_name)
    ax = degree.plot.bar(**kwargs)
    ax.set_xlabel('')
    return ax

degree_dict = dict(worldnet.degree(worldnet.nodes()))
nx.set_node_attributes(worldnet, degree_dict, 'degree')
top_degree = top_n(degree_dict, 20)
plot_top(top_degree)
```

```python
betweenness_dict = nx.betweenness_centrality(worldnet, weight='weight') # Run betweenness centrality
nx.set_node_attributes(worldnet, betweenness_dict, 'betweenness')
ax = plot_top(top_n(betweenness_dict, 14), title='Betweenness centrality', figsize=(12, 5),
              legend=False,
             rot=45 ,
              ylabel='Betweeness centrality',
             )
```

# How dependent is one particular countries on others?

```python
from typing import Optional
from typing import Dict
from sklearn.preprocessing import normalize

def betweenness (alpha2) -> float:
    return worldnet.nodes[alpha2]['betweenness']

def relative_reliance (alpha2) -> Optional[Dict[str, float]]:
    '''Returns betweeness centrality of nodes' neighbors'''
    connected_to =  [e[1] for e in worldnet.edges(alpha2)]
    if len(connected_to)>0:
        betweennesses = np.array([ betweenness(cc) for cc in connected_to ])
        normalized_betweennesses = normalize([betweennesses])[0]
        return dict(zip(connected_to, normalized_betweennesses))
    
relative_reliance('IR')    
```

# Create risk profiles for particular countries

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable
def profile (alpha2):
    scaler = 2
    fig, axs = plt.subplots(2, gridspec_kw={'height_ratios': [6, 2]},
                            figsize=(6*scaler, 4*scaler),
                            dpi=200,
                           )
    fig.suptitle(country_alpha2_to_country_name(alpha2))
    
    rrs = relative_reliance(alpha2)
    myrel_df = pd.DataFrame({'iso_a3': [to_alpha3(cc) for cc in rrs.keys()],
                              'reliance': rrs.values()})
    
    worldmap_myrel = worldmap.merge(myrel_df, on='iso_a3', how='left')

    # make legnd align
    divider = make_axes_locatable(axs[0])
    cax = divider.append_axes("right", size="1%", pad=0.1)
    
    # hide ticks on axes
    axs[0].axes.yaxis.set_visible(False)
    axs[0].axes.xaxis.set_visible(False)
    
    # plot reliances
    plot(worldmap_myrel, 'reliance', ax=axs[0], legend=True, cax=cax,
         legend_kwds={'label': "Relative reliance",}, # label  legend
         missing_kwds={'color': 'lightgrey'},  # disconnected countries should be gray
         vmin=0, vmax=1, # set color scheme to go from 0 to 1
        )
#                        'orientation': "horizontal"})

    # highlight country in question
    my_country = worldmap[worldmap['iso_a3']==to_alpha3(alpha2)]
    plot(my_country,  ax=axs[0], color='gray', hatch='///', edgecolor='red', aspect='equal')
    
    reliances = worldmap_myrel[worldmap_myrel['reliance']>0][['name', 'reliance']]\
        .sort_values(by='reliance', ascending=False)
    reliances.plot.bar(x='name', rot=90, ax=axs[1], legend=False)
    axs[1].yaxis.grid()
    
profile('US')
```

## risk metric

gini coefficient (i.e., inequality of relative dependence) divided by degree

```python
def gini(array):
    """Calculate the Gini coefficient of a numpy array."""
    # based on bottom eq: http://www.statsdirect.com/help/content/image/stat0207_wmf.gif
    # from: http://www.statsdirect.com/help/default.htm#nonparametric_methods/gini.htm
    # via: https://github.com/oliviaguest/gini/blob/master/gini.py
    array = array.flatten() #all values are treated equally, arrays must be 1d
    if np.amin(array) < 0:
        array -= np.amin(array) #values cannot be negative
    array += 0.0000001 #values cannot be 0
    array = np.sort(array) #values must be sorted
    index = np.arange(1,array.shape[0]+1) #index per array element
    n = array.shape[0]#number of array elements
    return ((np.sum((2 * index - n  - 1) * array)) / (n * np.sum(array))) #Gini coefficient
```

```python
def safety_metric (alpha2: str) -> float:
    '''
    metric of safety from cable cutting attack
    '''
    dependencies =  np.array(list(relative_reliance(alpha2).values()))
    return  len(dependencies) * (1/gini(dependencies))

risk_metric('US')
```

```python
countries = []
metrics = []
for node in worldnet.nodes():
    try:
        r = risk_metric(node)
#         c = country_alpha2_to_country_name(node) 
        countries.append(node)
        metrics.append(r)
    except AttributeError:
        pass
    except KeyError:
        pass

safety = pd.DataFrame({'country': countries, 'safety': metrics})\
            .sort_values(by='safety', ascending=False)

safety.head(20).plot.bar(x='country')
# safety.tail(20).plot.bar(x='country')
```

### marshall into polisci-vis format

```python
import json
with open('visualization/countries_codes_and_coordinates.json', 'r') as f:
    vis_ccs = json.load(f)
safe_alpha3s = set([ v['Alpha-3 code'] for v in vis_ccs['countryData']])
```

```python
#  https://github.com/lilybhattacharjee5/polisci-vis
polisci_vis_data = {}
for node in worldnet.nodes:
    rr = relative_reliance(node)
    if rr:
        for connection, reliance in rr.items():
            try:
                from_cc = to_alpha3(node)
                to_cc = to_alpha3(connection)
            except:
                print('Cannot find one of these CCs', node, connection)
                pass
            if from_cc in safe_alpha3s and to_cc in safe_alpha3s:
                k = f'{from_cc}->{to_cc}'
                v = {
                    'Overall_Similarity':  reliance*100
                }
                polisci_vis_data[k] = v
            else:
                print('skipping', node, connection)
```

```python
import json
with open('visualization/data.json', 'w') as f:
    json.dump(polisci_vis_data, f)
```
