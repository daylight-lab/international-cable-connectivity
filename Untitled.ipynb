```python
%pylab inline
import networkx as nx
```

```python
undersea = nx.read_gpickle('data/undersea_today.networkx')
terrestrial = nx.read_gpickle('data/terrestrial_today.networkx')
```

```python
world = nx.compose(undersea,terrestrial)
```

```python
nx.draw(world, with_labels=True)
```

```python
# Next, use nx.connected_components to get the list of components,
# then use the max() command to find the largest one:
components = nx.connected_components(world)
largest_component = max(components, key=len)

# Create a "subgraph" of just the largest component
# Then calculate the diameter of the subgraph, just like you did with density.
#

subgraph = world.subgraph(largest_component)
diameter = nx.diameter(subgraph)
diameter
```

```python
from pycountry_convert import country_alpha2_to_country_name
from operator import itemgetter
import pandas as pd

def top_n (my_dict, n, reverse=True):
    sorted_dict = sorted(my_dict.items(), key=itemgetter(1), reverse=reverse)
    return sorted_dict[:n]

def plot_top (top_lst, title='Centrality', **kwargs):
    degree = pd.DataFrame(top_lst, columns=['Country', title])
    degree.index = degree['Country']
    degree['Country'] = degree['Country'].apply(country_alpha2_to_country_name)
    ax = degree.plot.bar(**kwargs)
    ax.set_xlabel('')
    return ax

degree_dict = dict(world.degree(world.nodes()))
nx.set_node_attributes(world, degree_dict, 'degree')
top_degree = top_n(degree_dict, 20)
plot_top(top_degree)
```

```python
betweenness_dict = nx.betweenness_centrality(world) # Run betweenness centrality
nx.set_node_attributes(world, betweenness_dict, 'betweenness')
ax = plot_top(top_n(betweenness_dict, 14), title='Betweenness centrality', figsize=(12, 5))
```

```python
lowest_degree = top_n(degree_dict, 30, reverse=False)
for cc2, degree in lowest_degree:
    if degree==1:
        try:
            from_country = country_alpha2_to_country_name(cc2)
        except:
            from_country = cc2
        edges = world.edges(cc2)
        for edge in edges:
            try:
                to_country = country_alpha2_to_country_name(edge[1])
            except:
                to_country = edge[1]
            print(from_country, 'only connects to', to_country)
```
